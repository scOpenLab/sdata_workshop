```python
import easy_vitessce as ev
import spatialdata as sd
import scanpy as sc
```

```python
# Load the object
sdata = sd.read_zarr("PROJECTS/WORKSHOP/ZARR2-MB299-ALIGNED.zarr")
sdata
```

```python
# Convert from multiscale to a single array
# See this open PR: https://github.com/vitessce/easy_vitessce/pull/63/commits/d62b6565e1456acd7b59447f82bd5bac994cafcf
sdata["morphology_mip"] = sdata["morphology_mip"].scale0.image
sdata["comet_image"] = sdata["comet_image"].scale0.image
```

```python
# Vitessce plot with SpatialData syntax
sdata.pl.render_images("morphology_mip", channel = "DAPI")\
    .pl.render_shapes("cell_boundaries", color="CD4", table_name="table")\
        .pl.show(coordinate_systems = "global")
```

```python
# Vitessce plot with Scanpy syntax
sc.plotting.violin(adata=sdata["table"], keys="CD4", groupby="region")
```
